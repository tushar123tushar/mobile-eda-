```python
data['warranty Year'] = data['warranty'].str.extract(' (\d+) ').astype(float)
```

```python
greater_than = data[data['warranty Year']>2].head(5)
result = greater_than.sort_values(by='warranty' , ascending=False)
result[['name' , 'warranty']]
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=result , x='warranty' , hue='name')
plt.title('Name and Warranty Hight warranty Year')
plt.grid()
plt.ylabel('count')
plt.xlabel('warranty')
plt.show()
```
